```python
X_ = X.astype('float64')
```

```python
X_
```

```python
size = int(len(X_) * 0.64)

train, test = X_[0:size], X_[size:len(X_)]

history = [x for x in train]

predictions = list()
```

```python
for t in range(len(test)):
    
    model = VARMAX(series, order=(1, 1))
    
    model_fit = model.fit()
    
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
    print('predcited=%f, expected=%f' % (yhat, obs))
```
